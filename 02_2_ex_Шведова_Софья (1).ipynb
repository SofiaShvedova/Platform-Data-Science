## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

Сначала открываем исходный файл и создаем новый файл для записи, потом вычисляем средний балл, а затем записываем данные с средним баллом в новый файл

Словарь

````python
gpa = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
````

````python
with open('grades.csv', 'r') as file, open('grades_gpa.csv', 'w') as output:
    for line in file:
        name, grade = line.strip().split(';')
        gpa = {'A': 4, 'B': 3, 'C': 2, 'D': 1}.get(grade.strip(), 0)

        output.write(f'{name}; {grade.strip()}; {gpa}\n')
````

Открываем файл

````python
# импортировать библиотеку csv
import csv

# инициализировать массив верхнего уровня
data_values = []

# откройте файл и добавьте строки в виде массивов в data_values
with open('grades.csv') as csvfile:
    filereader = csv.reader(csvfile, delimiter=';', quotechar='"')
    for row in filereader:
        # Добавление среднего балла
        row.append(gpa[row[1].strip()])
        data_values.append(row)
````

Создадим датафрейм и запишем в новый файл данных в новый файл

````python
ocenki = pd.DataFrame(data_values, columns=['Ученик', 'Оценка буквой', 'Оценка числом'])
ocenki.to_csv('grades_gpa.csv', index=False)
````
