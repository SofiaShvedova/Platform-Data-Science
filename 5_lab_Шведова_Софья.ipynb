# 5.1

Реализовать графики в Altair с использованием следующих ключей:
10 вариант

Метки: mark_text(), mark_rect().

Столбцы данных: , 'Weight_in_lbs', 'Year'.

Кодировки: shape, row, column.

С помощью mark_rect можно создавать прямоугольные области на графике, которые могут быть закрашены определенным цветом или заполнены текстурой


Метка mark_rect() и кодировка shape

```python
alt.Chart(cars).mark_rect().encode(
    x=alt.X('Weight_in_lbs', bin=True),
    y=alt.Y('Year', bin=True),
    color='mean(Weight_in_lbs)',
    shape = 'Origin'
)
```

Метка mark_rect() и кодировка row

```python
alt.Chart(cars).mark_rect().encode(
    x=alt.X('Weight_in_lbs', bin=True),
    y=alt.Y('Year', bin=True),
    color='mean(Weight_in_lbs)',
    row = 'Origin'
)
```

Метка mark_rect() и кодировка column

```python
alt.Chart(cars).mark_rect().encode(
    x=alt.X('Weight_in_lbs', bin=True),
    y=alt.Y('Year', bin=True),
    color='mean(Weight_in_lbs)',
    column = 'Origin'
)
```

Mark_text позволяет добавлять текстовые надписи к графику, указывая координаты и содержание текста.


Метка mark_text() и кодировка shape

```python
alt.Chart(cars).mark_text().encode(
    x='Year:O',
    y='Weight_in_lbs:Q',
    text='Weight_in_lbs',
    shape='Weight_in_lbs:O'
).properties(width=1000, height=600)
```

Метка mark_text() и кодировка row

```python
alt.Chart(cars).mark_text().encode(
    x='Year:O',
    y='Weight_in_lbs:Q',
    text='Weight_in_lbs:Q',
    row='Year:N'
).properties(width=600, height=400)
```

Метка mark_text() и кодировка column

```python
alt.Chart(cars).mark_text().encode(
    x='Year:O',
    y='Weight_in_lbs:Q',
    text='Weight_in_lbs:Q',
    column='Year:N'
).properties(width=600, height=400)
```

# 5.2

Используя данные об автомобилях, создается диаграмма рассеяния (*scatter-plot*), на которой *размер* (*size*) точек становится больше при наведении на них курсора.

```python
import altair as alt

selection = alt.selection_single(on='mouseover', nearest=True)
chart = alt.Chart(cars).mark_circle().encode(
    x='Year:Q',
    y='Weight_in_lbs:Q',
    opacity=alt.value(0.5),
    tooltip=['Year', 'Weight_in_lbs'],
    color='Origin',
    size=alt.condition(
        selection,
        alt.value(200),
        alt.value(30)
    )
).add_selection(
    selection
).interactive()

chart
```

# 5.3

Используя данные об автомобилях, создается двухпанельная (*two-panel*) гистограмма (количество миль на галлон на одной панели, количество лошадиных сил на другой), где можно перетащить мышь, чтобы выбрать данные на левой панели, чтобы отфильтровать данные на второй панели.

```python
auto = alt.selection_interval(encodings=['x'])

his_m = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True),
    y='count()',
    color=alt.condition(auto, alt.value('red'), alt.value('gray'))
).add_selection(
    auto
)

his_h = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Horsepower:Q', bin=True),
    y='count()',
    color=alt.condition(auto, alt.value('orange'), alt.value('gray'))
).transform_filter(
    auto
)

his_m & his_h
```

Можно перетащить мышь, чтобы выбрать данные на левой панели, чтобы отфильтровать данные на второй панели.

# 5.4

Изменить приведенный выше пример диаграммы разброса и гистограммы, чтобы

- панорамировать и увеличивать диаграмму рассеяния;
- гистограмма отражала только те точки, которые видны в данный момент.

```python
scatterplot = alt.Chart(cars).mark_circle().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color=alt.condition(auto, alt.value('red'), alt.value('blue'))
).add_selection(
    auto
).properties(
    width=400,
    height=400
).interactive()

his_m = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True),
    y='count()',
    color=alt.condition(auto, alt.value('orange'), alt.value('red'))
).transform_filter(
    auto
).properties(
    width=400,
    height=400
)

(his_m | scatterplot)
```

Как можно заметить, что если изменять из рассеянной диаграммы диапазон, то столбчатая диаграмма меняет свой вид

# 5.5

Создание диаграммы на основе  сгенерированных данных и постройка кривые синуса и косинуса с помощью `transform_calculate`
Вариант 10 (по списку в успеваемости):
x = pd.DataFrame({'x': np.linspace(-(последняя цифра группы), (первая цифра группы))})
Первая цифра группы - 2 (Адэу-211)
Последняя цифра группы - 1 (Адэу-211)

```python
import pandas as pd
import numpy as np

x = pd.DataFrame({'x': np.linspace(-2, 1)})

x['sin'] = np.sin(x['x'])
x['cos'] = np.cos(x['x'])

ch = alt.Chart(x).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'

).mark_line().encode(
    x='x',
    y='sin',
    color=alt.value('blue')
) + alt.Chart(x).mark_line().encode(
    x='x',
    y='cos',
    color=alt.value('red')
)
ch.transform_filter("datum.cos >= datum.sin")
ch
```

# 5.7

Настройка эстетики предыдущего графика

```python
chart = alt.Chart(x).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'

).mark_line().encode(
    x=alt.X('x', axis=alt.Axis(title='Значения x', labels=True, ticks=True)),
    y=alt.Y('sin', axis=alt.Axis(title='Значения sin(x)', labels=True, ticks=True), scale=alt.Scale(domain=[-1, 1])), # измение  пределов `x` и `y`.
    color=alt.value('blue'),
    strokeWidth=alt.condition("cos(datum.x)", alt.value(5), alt.value(1)) # изменеие толщины линии
) + alt.Chart(x).mark_line().encode(
    x='x',
    y='cos',
    color=alt.value('red'),
    strokeWidth=alt.condition("sin(datum.x)", alt.value(5), alt.value(1)),

)

chart = chart.properties(
    title='График синуса и косинуса',
    width=500,
    height=300
)

chart
```
